```python
import numpy as np
import pandas as pd
import seaborn as sns

from collections import defaultdict
from glob import glob
```

```python
classifiers = ['ExtraTreeClassifier',
 'DecisionTreeClassifier',
 'MLPClassifier',
 'KNeighborsClassifier',
 'SGDClassifier',
 'RidgeClassifierCV',
 'RidgeClassifier',
 'PassiveAggressiveClassifier',
 'GaussianProcessClassifier',
 'AdaBoostClassifier',
 'GradientBoostingClassifier',
 'BaggingClassifier',
 'RandomForestClassifier',
 'BernoulliNB',
 'CalibratedClassifierCV',
 'GaussianNB',
 'LabelPropagation',
 'LabelSpreading',
 'LinearDiscriminantAnalysis',
 'LinearSVC',
 'LogisticRegression',
 'LogisticRegressionCV',
 'NearestCentroid',
 'Perceptron',
 'QuadraticDiscriminantAnalysis',
 'SVC']
```

```python
from math import sqrt, exp
df = defaultdict(list)
files = sorted([f for f in \
                glob('../experiments_cell/RESAMPLED48_CELL*.sh.o*')+\
                glob('../experiments_cell/FINE_CELL*.sh.o*')+\
                glob('../experiments_cell/*CPS*.sh.o*')     

for f in files:   
    if 'RESAMPLED' in f:       
        sample_size = int(f.split('/')[-1].split('_')[0].lstrip('RESAMPLED'))*3
    else:
        sample_size = 713
                          
               
    if "_SYNTHETIC" in f or "_CPS_" in f:
        if sample_size == 48*3:                 
            data_source = 'Uniform noise (144 cells, balanced)'
        else:
            data_source = 'Uniform noise (713 cells, imbalanced)' 
    else:
        if sample_size == 48*3:               
            data_source = 'Uninformative genes (144 cells, balanced)'
        else:
            data_source = 'Uninformative genes (713 cells, imbalanced)'  
#         data_source = 'Synthetic'
    with open(f) as file: 
        data = file.read()
        lines = data.split(f"[{sample_size},")[1:]
        for line in lines:
            raw = line.split("test_balanced_accuracy': array(")   
            temp = [sample_size, int(line.split(",")[0]), 3,0]
            for j in range(5):
                df['Ref'].append((temp[2],temp[3]))
                df['Data'].append(""+data_source)
                df["Number of samples"].append(0+sample_size)
                df["Number of uninformative features"].append(temp[1]+0)
                df["Signal/Noise ratio"].append(round((1+temp[3])/sqrt(0.1),1)/(temp[1]+10**-64))
                df["Feature/Dist ratio"].append((temp[1])/exp((round((1+temp[3])/sqrt(0.1),1))))
                df["Number of clusters"].append(temp[2]+0)
                df["Distance between clusters"].append(1+temp[3])
                df["Separation between clusters"].append(round((1+temp[3])/sqrt(0.1),1))
                for i, cls in enumerate(classifiers):
                    acc = eval(raw[1+i].split(")")[0])
                    df[cls].append(acc[j])                   
data = pd.DataFrame.from_dict(df)
data
```

```python
alpha=0.6
classes_list = [['KNeighborsClassifier','RidgeClassifier','SVC',  'GradientBoostingClassifier']]
palette = [
    "#fdae61",
    "#d7191c",
     "#abdda4",
    "#2b83ba",
]
hue_order = sorted(data.Data.unique())
fig,ax =plt.subplots(len(classes_list),4,figsize=(xs*2,ys*0.8),sharex=True)
for nn, classes in enumerate(classes_list):
    for i, cls in enumerate(classes):     
        df = data[["Number of uninformative features",'Number of samples','Data',cls]].sort_values(by="Number of uninformative features")
        df["Number of uninformative features"] =  3*(df["Number of uninformative features"])/df['Number of samples']
        df['Relative loss of accuracy'] = 100*(1- df[cls])/(1-1/3)
        sns.lineplot(x="Number of uninformative features", hue='Data',style='Data',palette=palette,
                     hue_order = hue_order,style_order = hue_order,
                     y='Relative loss of accuracy',markers=True,data=df,linewidth=4,legend=(i==3), ax = ax[i],
                     err_style="band",ci='sd',err_kws={'linewidth':2})   
        if i==3:
            handles, labels = ax[i].get_legend_handles_labels()
            ax[i].get_legend().remove()
        custom_frame(ax[i])
        ax[i].set_ylabel('Loss of predictability (\%)',fontsize=30)
        ax[i].set_xlabel("")
        ax[i].set_title(f"{cls}",fontsize=30,position=(0.5, 1))
        ax[i].set_ylim(-10,110)
        ax[i].set_yticks([0,25,50,75,100])
        if nn!=len(classes_list)-1:
            ax[i].set_xlabel('') 
            ax[i].set_xticklabels([])
            ax[i].tick_params(labelbottom=False)
        elif i!=1: ax[i].set_xlabel('') 
                
        if i!=0:
            ax[i].set_ylabel('') 
            ax[i].set_yticklabels([])
            ax[i].tick_params(labelleft=False)
ax[i].set_xscale('log')
ax[i].set_xticks([0.01,0.1,1,10,100])
plt.subplots_adjust(wspace=0.1,hspace=0.3)
fig.text(0.5, -0.05, 'Number of uninformative features/sample size per cluster', ha='center',fontsize=30)

fig.legend(handles, 
           labels, 
           bbox_to_anchor = (0.35,1.41),
           loc='upper left', 
           ncol=2, 
           title='Data',
           title_fontsize=30,
           columnspacing=0.5,
           frameon=False)
plt.show() 
```
