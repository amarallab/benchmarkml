```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
classifiers = pd.read_pickle('../data/multi_clusters_algorithms_compare_1.pkl').columns.tolist()

acc0 = {
    2:0.943,
    3:0.920,
    4:0.886,
    5:0.884,
    6:0.867,
    7:0.868,
    8:0.856,
    9:0.848,
}
```

```python
from math import sqrt
from collections import defaultdict
from glob import glob
# classifiers = cls.copy()
df = defaultdict(list)

files = [f for f in glob('../quest/fullex_G*.sh.o333*') if 3335563<=int(f[-7:])<=3335662]
for f in files:
    with open(f) as file: 
        for line in file:
            line = line.lstrip('[').rstrip(']\n')
            data = line.split(', array')
            temp = eval(data[0])
            if temp[3]!=0: continue
            if temp[0]!=1000: continue
            if temp[1] not in set([0,2,20,50,100,1000,5000,8000]): continue
            df['Ref'].append((temp[1],temp[2]))
            df["Number of samples"].append(temp[0])
            df["Number of uninformative features"].append(temp[1])
            df["Number of clusters"].append(temp[2])
            df["Distance between clusters"].append(1+temp[3])
            df["Separation between clusters"].append(round((1+temp[3])/sqrt(0.1),1))
            for i, c in enumerate(classifiers):
                df[c].append(np.mean(eval(data[i+1])))
    
data = pd.DataFrame.from_dict(df)
data = data.groupby('Ref').mean()
df = data.copy()
for col in classifiers:
    df[col] = data.apply(lambda row:(acc0[row['Number of clusters']]-row[col])/acc0[row['Number of clusters']]*100,axis=1)
df = df.groupby('Ref').mean()
df_mean = df[classifiers]
df_mean2 = data.groupby('Ref').mean()[classifiers]
```

```python
rla_cls =['KNeighborsClassifier', 'GaussianProcessClassifier','NearestCentroid',
    'RidgeClassifier','Perceptron','LogisticRegressionCV',
    'MLPClassifier', 'RandomForestClassifier','LogisticRegression', 
     'SVC', 'LinearDiscriminantAnalysis', 'GaussianNB',
    'DecisionTreeClassifier', 'GradientBoostingClassifier', 'BaggingClassifier']
# print(len(rla_cls))
df = pd.read_pickle('../data/multi_clusters_algorithms_compare.pkl').reset_index()
df['Number of clusters'] = df['index'].apply(lambda x:x[1])
df = df[df['Number of clusters']<10]
df['Acc0'] = df['index'].apply(lambda x:acc0[int(x[1])])
df = df.set_index('index').sort_index()

for col in df.columns:
    df[col] = df.apply(lambda row:(row['Acc0']-row[col])/(row['Acc0']-1/row['Number of clusters'])*100,axis=1)
df = df.drop(columns=['Number of clusters','Acc0'])
df_mean = df.applymap(np.mean)
stdrcparams(usetex=True)
fig, ax=plt.subplots(figsize=(xs*2,ys*1.5))
sns.set(font_scale=1.2)
g = sns.heatmap(df_mean[rla_cls].T,ax=ax,
                   xticklabels=2,yticklabels=1, cmap = "Blues",cbar=False,)
#                    cbar_ax=cbar_ax)
#                 cbar_kws={'label': 'Relative loss of predictability (\%)','use_gridspec':False,'location':'bottom','shrink':0.3})
ax.set_xlabel('(Number of uninformative features, Number of clusters)',fontsize=30)
ax.tick_params(labelleft=False,labelright=True)
plt.yticks(rotation='horizontal')
plt.tight_layout()
plt.savefig('../Figs/NULL_SUB_Fig2A_RLP.pdf',bbox_inches='tight',transparent=True)
plt.savefig('../Figs/NULL_SUB_Fig2A_RLP.png',bbox_inches='tight',transparent=True)
plt.show()
```
