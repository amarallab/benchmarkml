```python
from math import sqrt,log
df = defaultdict(list)
files = [f for f in glob('../experiments/Sample_G*.sh.o*')]
for f in files:
    with open(f) as file: 
        for line in file:
            line = line.lstrip('[').rstrip(']\n')
            data = line.split(', array')
            temp = eval(data[0])
            for j in range(5):
                df['Ref'].append((temp[2],temp[3]))
                df["Number of samples"].append(int(temp[0])*0.8)
                df["Number of uninformative features"].append(temp[1])
                df["Sample/Feature ratio"].append(int(temp[0])*0.8/(int(temp[1])+10**-32))
                df["Feature/Sample ratio"].append((int(temp[1])/temp[0])*0.8)
                df["Dimension/Sample ratio"].append((int(temp[1]+2)/temp[0])*0.8)
                df["Number of clusters"].append(temp[2])
                df["Distance between clusters"].append(1+temp[3])
                df["Separation between clusters"].append(round((1+temp[3])/sqrt(0.1),1))
                for i, cls in enumerate(classifiers):
                    df[cls].append(eval(data[i+1])[j])
data = pd.DataFrame.from_dict(df)
data.head()
```

```python
classes_list = [['KNeighborsClassifier', 'GaussianProcessClassifier','NearestCentroid',],
    [ 'RidgeClassifier','Perceptron','LogisticRegressionCV'],
    ['MLPClassifier', 'RandomForestClassifier','LogisticRegression',], 
     [ 'SVC', 'LinearDiscriminantAnalysis', 'GaussianNB'],
    ['DecisionTreeClassifier', 'GradientBoostingClassifier', 'BaggingClassifier'],
]
palette1 = ['#fff5eb',
'#fee6ce',
'#fdd0a2',
'#fdae6b',
'#fd8d3c',
'#f16913',
'#d94801',
'#8c2d04']
palette2 = ['#f7fbff',
'#deebf7',
'#c6dbef',
'#9ecae1',
'#6baed6',
'#4292c6',
'#2171b5',
'#084594']
palette3 = [
'#f7f4f9',
'#d4b9da',
'#c994c7',
'#df65b0',
'#e7298a',
'#ce1256',
'#91003f',
]

from scipy.optimize import curve_fit
def f(x, a, b):
    return 1 / (1 + np.exp(-a * x + b))

def g(x, a, b):
    return (a*np.exp(b-a*x))/ (1 + np.exp(-a * x + b))**2

fig,ax =plt.subplots(len(classes_list),3,figsize=(xs*1.5,ys*len(classes_list)/2))
for nn, classes in enumerate(classes_list):
    for i, cls in enumerate(classes):     
        df = data.copy()
        df=df.explode(cls).reset_index()
        df[cls] = df[cls].astype(float)
        df['Relative loss of accuracy'] = df.apply(lambda row: 
                  (acc0[row['Distance between clusters']][row['Number of clusters']]
                   -row[cls])/(acc0[row['Distance between clusters']][row['Number of clusters']]-1/row['Number of clusters']),axis=1)

        df = df[["Number of uninformative features",
                 'Relative loss of accuracy',
                 'Dimension/Sample ratio',
                 "Separation between clusters",
                 cls]].reset_index(drop=True)
        df = df.groupby(by=['Number of uninformative features','Dimension/Sample ratio']).mean().reset_index()
   
        df = df[10**-9<=df['Dimension/Sample ratio']]
        xdata = np.log10(df["Dimension/Sample ratio"].values)
        ydata = df["Relative loss of accuracy"].values
        p0 = [1/2, 1]  # initial parameter estimate
        popt, pcov = curve_fit(f, xdata, ydata)

        ax[nn,i].scatter(xdata, ydata*100,facecolors='none',edgecolors='blue')
        ax[nn,i].plot(np.linspace(-3,4,100),[100*f(i, popt[0], popt[1]) for i in np.linspace(-3,4,100)],'r--',linewidth =2)
        ax[nn,i].plot(np.linspace(-3,4,100),[100*g(i, popt[0], popt[1]) for i in np.linspace(-3,4,100)],'g--',linewidth =2)

        res = 0
        curr = float('-inf')
        for x in np.linspace(-3,4,100):
            if g(x, popt[0], popt[1]) > curr:
                curr = g(x, popt[0], popt[1])
                res = x     
        if res != -3:
            ax[nn,i].vlines(res,0, 1*100,colors='green',linestyles=':',linewidth =2)
            ax[nn,i].text(res,1*100,f"{res:0.2f}", fontsize=15)
        custom_frame(ax[nn,i])
        ax[nn,i].set_xticks([-2,0,2,4])
        ax[nn,i].set_ylim(-20,120)
        ax[nn,i].set_yticks([0,25,50,75,100,125])
        ax[nn,i].set_ylabel('Loss of predictability (\%)',fontsize=40)
        ax[nn,i].set_xlabel('Log10 (number of Dimension/sample size per cluster)',fontsize=40)
        ax[nn,i].set_title(f"{cls}",fontsize=20,position=(0.5, 0.9))
        if nn!=len(classes_list)-1:
            ax[nn,i].set_xlabel('') 
            ax[nn,i].set_xticklabels([])
            ax[nn,i].tick_params(labelbottom=False)
        elif i!=1: ax[nn,i].set_xlabel('') 
                
        if i!=0:
            ax[nn,i].set_ylabel('') 
            ax[nn,i].set_yticklabels([])
            ax[nn,i].tick_params(labelleft=False)
        elif nn!=2:
            ax[nn,i].set_ylabel('') 
plt.subplots_adjust(wspace=0.1,hspace=0.3)
plt.show() 
```
